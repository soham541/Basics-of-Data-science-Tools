#### Figures and subplots

```python
x = np.linspace(1,10,100)
y = np.log(x)
plt.figure(1)

# Equivalend to 1,2,1 [1,2] 1st element
plt.subplot(121)
plt.title("y = log(x)")
plt.plot(x,y)

# Second subplot
plt.subplot(122)
plt.title("y = log(x)**2")
plt.plot(x,y**2)
```
