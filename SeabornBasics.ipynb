```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
df = pd.read_csv("all_stocks_5yr.csv")
```

```python
df.head()
```

```python
sns.displot(df['high'], rug=True)
```

```python
sns.distplot(df['low'], hist=False)
```

```python
plt.subplot(2,2,1)
plt.title('Opening')
sns.distplot(df['open'])
```

```python
sns.boxplot(df['open'])
```

```python
sns.jointplot('high','low',df)
```

```python
df = df[(df.open < 10) & (df.low > 15)]
```

```python
sns.jointplot('open','low',df)
```

```python
grp = df['open','low','volume']
```

```python
sns.pairplot(grp)
```
